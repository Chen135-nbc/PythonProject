### 2.双分支（基于判断实现：

```python
if 判断条件:
    条件成立时执行的代码1
    条件成立时执行的代码2
    ……
else:
    条件不成立时执行的代码1
    条件不成立时执行的代码2
    ……
```


````python
age = int(input("请输入你的年龄："))
if age >= 18:
    print("你是成年人")
    print("你能考驾照了")
else:
    print("你是未成年人")
    print("你还不能考驾照")
````

### 3.多分支

>各分支之间时互斥的

````python
age = int(input("请输入你的年龄："))
if age <= 10:
    print("你是儿童")
elif age <= 18:
    print("你是青少年")
elif age <= 30:
    print("你是青年")
elif age <= 50:
    print("你是中年")
elif age <=70:
    print("你是中老年")
else:
    print("你是老年")
````

## 循环

- while循环
- for 循环

### while循环
```python
while 循环条件:
    条件成立时执行的操作1
    条件成立时执行的操作2
```

````python
n = 1
while n <= 10:
    print(f'第{n}次你好呀')
    n += 1
````

### for循环





**for循环语法格式**：
``` python
for 当前取出的值 in 可迭代对象:
    要执行的循环操作1
    要执行的循环操作2
    ……
```

**注意点**

1. 可迭代对象中的元素有多少个，循环就执行多少次
2. 只要不在循环内部修改可迭代对象，for循环通常不会出现死循环

````python
text = input("请输入要加密的文字：")
secret = ''

for t in text:
    unicode = ord(t)
    secret += chr(unicode + 1)
print(f'经过加密的内容为{secret}')
````

### 对比while循环和for循环

| while | for |
|-------|-----|
| 条件驱动循环（只要条件为True就继续循环）| 迭代驱动循环（遍历一个可迭代对象的每一项）|
| 写法灵活，尤其适合处理不确定循环次数的场景 | 语法简洁，处理迭代对象的首选，不易发生死循环 |
| 需手动维护条件，可能容易导致死循环 | 知识和一开始就知道要循环几次的场景 |

### 嵌套循环

````python
day = 1
for day in range(1,31):
    print(f'********* 第{day}天 **********')
    for group in range(1,4):
        print(f'这是第{group}组仰卧起坐')
    print(f'第{day} 天任务已完成！明天继续！\n')
print(f'为期{day}天的健身计划已完成')
````

````python
# 九九乘法表
for row in range(1,10):
    for item in range(1, row + 1):
        print(f'{item}*{row}={item * row}', end = '\t')
    print()
````

### continue和break
- continue: 跳过本次循环剩余语句，随后进入下一次循环
- break：立即终止循环，不再执行后续循环
