## 01-字面量

## 02-变量

- 在Python中，变量就是一个代号，用来和某个值建立绑定关系
- 语法: `变量名 = 值`
- 在Python中，当一个新的变量出现时，必须立即与某个值建立绑定关系-> 赋值操作
- 变量名要遵守一定的命名规则


```python
name = "Nancy"
age = 21
weight = 47.8

print(name,"的年龄是",age,'她的体重是',weight)
```

## 03-标识符明明规则

标识符：程序中我们可以自己起名的内容(变量名、函数名、类名、模块名……)

**标识符的命名规则**
1. 只能包含：数字、字母、下划线，且**不能以数字开头**，**不能包含空格**。
2. 标识符**区分大小写**
3. 标识符不能使用关键字
4. 标识符尽量不要与内置函数同名
5. 追求：简洁清晰，具有描述性

**命名方式**
1. 大驼峰：`StudentScore`
2. 小驼峰：`studentScore`
3. 蛇形： `student_score`

## 09-字符串

### 四种定义方式

- 单引号和双引号的写法是等价的，不能直接换行，需要用圆括号才能换行

- 三个单引号的写法，可以直接换行，并且可以作为多行注释使用

- 三个双引号的写法，可以直接换行，也可以作为多行注释使用，还能作为文档字符串使用

### 格式化输出
- 写法1: 直接用加号进行拼接


- 写法2: 使用占位符
  >%s占位字符串，%f占位浮点数，%i占位整数，%d占位十进制的整数，%s是万能的


- 写法3: 使用f-string， 目前python最推荐的方式


```python
name = 'Nancy'
gender = '女'
weight = 47.8
age = 21

info1 = '我叫' + name + '我是' + gender + '生'
print(info1)

info2 = '我叫%s,我是%s生,我的体重是%f,年龄是%i' % (name,gender,weight,age)
print(info2)

info3 = f'我叫{name},我是{gender}生,我的体重是{weight},年龄是{age}'
print(info3)
```

### 字符串占位符精度控制

#### %m.ns 字符串占位
m -> 
1. 字符串的最小宽度，位数不够会使用空格补全，位数小于字符串长度则不起作用
2. 正数是右对齐，负数是左对齐
n ->
精度控制，最多输出n个字符（若n大于实际字符串长度，则不起作用）

#### %m.nf 浮点数占位
m -> 
1. 控制整体宽度（整体宽度 = 整体宽度 + 小数点 + 小数宽度）
2. 位数不够空格来补，位数小于整体宽度，则自动失效
3. 正数是右对齐，负数是左对齐
n ->
精度控制，保留n位小数（n默认值为6),不够0来补，截断时会四舍五入

#### %m.nd 十进制数占位
同理

### 4.逻辑运算符

| 运算符 | 名称 | 功能描述 |
|----|---- |----|
| and | 逻辑与| 判断其两侧的值，是否都为True |
| or | 逻辑或 | 判断其两侧的值，是否至少有一个为True |
| not | 逻辑非 | 用于对一个值取反|


#### and

- and用于判断其两侧的值，是否都为True

- and具备“逻辑短路”能力
print(False and 3 / 0)

- and返回的不一定是布尔值，它返回的是某个参与计算的值本身
  
  - 规则： and会先判断左边，如果左边为False，这直接返回左边的值，否则返回右边的值

  - 备注：若参与and运算的值不是布尔值，Python会自动转为布尔值后进行逻辑操作


```python
# and用于判断其两侧的值，是否都为True
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
# and具备“逻辑短路”能力
print(False and 3 / 0)
```

```python
# and返回的不一定是布尔值，它返回的是某个参与计算的值本身
print(2 - 2 and True)
print('' and True)
print(True and 4 / 2)
print( 3 + 3 and 3 * 4)
```

#### or

- or用于判断其两侧，是否至少有一个为True -> **只要有一个是True，就返回True**

- or同样具备“逻辑短路”能力

- or返回的不一定是布尔值，它返回的是某个参与计算的值本身
  
   - 规则： or会先判断左边，如果左边为"真"，这直接返回左边的值，否则返回右边的值
  
   - 备注：若参与or运算的值不是布尔值，Python会自动转为布尔值后进行逻辑操作


```python
# or用于判断其两侧，是否至少有一个为True -> 只要有一个是True，就返回True
print(9 > 2 or 9 > 2)
print(9 > 2 or False)
print(False or 9 > 2)
print(False or False)
```

```python
# or返回的不一定是布尔值，它返回的是某个参与计算的值本身
print(7 - 2 or False)
print('你好' or 'Nancy')
print(False or 8 / 2)
print(2 - 2 or 3 * 4)
```

#### not用于取反
- 备注：若参与not运算的值不是布尔值，Python会自动转为布尔值后进行逻辑操作

- not返回的值一定是布尔值
  


```python
#若参与not运算的值不是布尔值，Python会自动转为布尔值后进行逻辑操作
print(not True)
print(not False)
print(not 3 > 2)
print(not 3 < 2)
```

## 12-进制

- 0b开头表示二进制
`num1 = 0b11001`

- 0o开头表示八进制
`num2 = 0o1034`

- 0x开头表示十六进制
`num3 = 0x1cf`

>Python中所有的非十进制数字，只是代码层面的编写方式，是给程序员看的

>Python在对上面的num1、num2、num3进行计算、打印等操作时，会自动将其转为十进制

### 进制转化

- 使用`bin()`将十进制转为二进制
`result1 = bin(25)`

- 使用`oct()`将十进制转为八进制
`result2 = oct(540)`

- 使用`hex()`将十进制转为十六进制
`result3 = hex(463)`

- 使用int()将指定进制的数，转为十进制数字
  
    `value1 = int('0b110011',2)`

    `value2 = int('0o10342',8)`

    `value3 = int('0x1cf6',16)`



```python
num1 = 0b11001
num2 = 0o1034
num3 = 0x1cf

# Python中所有的非十进制数字，只是代码层面的编写方式，是给程序员看的
# Python在对上面的num1、num2、num3进行计算、打印等操作时，会自动将其转为十进制
print(num1,num2,num3)
print(num1 + 2)
print(str(num2))
print(num3 > 400)
```

```python
# - 使用bin()将十进制转为二进制
result1 = bin(25)
# - 使用oct()将十进制转为八进制
result2 = oct(540)
# - 使用hex()将十进制转为十六进制
result3 = hex(463)
print(result1)
print(result2)
print(result3)
```

```python
# 使用int()将指定进制的数，转为十进制数字
value1 = int('0b110011',2)
value2 = int('0o10342',8)
value3 = int('0x1cf6',16)
print(value1,value2,value3)
```
