# 第五章-函数

## 01-函数的概念及分类

>官方定义：函数(function)是组织好的，可重复使用的、用于执行特定任务的代码块。

### 函数的作用
- 代码复用
- 结构清晰
- 模块化开发

### 函数的分类

| 分类 | 说明 |
| ----| -----|
| 内置函数| 无需要任何操作，直接就能够使用 |
| 模块提供的函数 | 需要导入只能的模块后，才能使用 |
| 自定义函数 | 程序员自己定义的函数（先定义再使用）|

## 03-参数的使用



>形参的使用范围仅限函数内


```python
def order(num,dish):
    print(f'您点的是：{num}份{dish}')

order(1,'糖醋里脊')
```

## 04-位置参数

- 位置参数：调用函数时，根据参数在函数定义中出现的顺序，把实参的值依次传递给对应的形参


>!!! 使用位置参数时，不能跳过某个位置的参数，去给后面的形参赋值

```python
# 定义函数（接受位置参数）
def greet(name,gender,age,height):
    print(f'我叫{name},性别{gender},年龄是{age},身高是{height}cm')

greet('Nancy','female',21,163)
```

## 05-关键字参数

- 调用函数时，通过`形参名=值` 的形式传递实参



> ！！！使用关键字参数时，位置参数必须在关键字参数之前

## 06-限制传参方式

**具体规则：`/`前面只能使用位置参数，`*`后面只能使用关键字参数。**



## 07-参数默认值

**定义函数时，通过`形参名=值` 的形式，为参数指定一个默认值**



## 08-可变函数

### 可变位置参数

**定义函数时，在形参名前加`*`, 可接受任意数量的位置参数，并打包成一个元组**



### 可变关键字参数

**定义函数时，在形参名前加`**` ，可以接受任意数量的关键字参数，并打包成一个字典**



> 可变位置参数、可变关键字参数，可以同时使用，但必须**先写可变位置参数**
