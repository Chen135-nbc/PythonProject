# 第三章·Python核心基础

## 01-字面量

1. 字面量： 直接写在代码中的**具体值**
2. Python的字符串中可以包含任意字符，必须使用引号包起来
3. 字符串的引号，可以是：单引号、双引号、三个单引号、三个双引号
4. 写在Python文件头部的字符串，会被自动识别为docstring(文档字符串)
5. 文档字符串的主要作用：对当前Python文件进行一些说明，且文档字符串必须用**三个双引号**

## 02-变量

- 在Python中，变量就是一个代号，用来和某个值建立绑定关系
- 语法: `变量名 = 值`
- 在Python中，当一个新的变量出现时，必须立即与某个值建立绑定关系-> 赋值操作
- 变量名要遵守一定的命名规则


```python
name = "Nancy"
age = 21
weight = 47.8

print(name,"的年龄是",age,'她的体重是',weight)
```

## 03-标识符明明规则

标识符：程序中我们可以自己起名的内容(变量名、函数名、类名、模块名……)

**标识符的命名规则**
1. 只能包含：数字、字母、下划线，且**不能以数字开头**，**不能包含空格**。
2. 标识符**区分大小写**
3. 标识符不能使用关键字
4. 标识符尽量不要与内置函数同名
5. 追求：简洁清晰，具有描述性

**命名方式**
1. 大驼峰：`StudentScore`
2. 小驼峰：`studentScore`
3. 蛇形： `student_score`

## 04-常量

1. 常量：一旦被赋值，就不希望被修改的量
2. 用**全大写变量名**来表示常量，涉及多个单词用**下划线**做分隔
3. `ctrl+shift+U` 切换大小写
4. 没有强制的常量机制，本质还是变量，但约定不去修改


## 05-注释

1. 注释：对代码的备注和解释，代码执行时通常不起作用
2. 注释的分类：
    1. 单行注释 `#` 后的一行内容，会被视为注释
    2. 多行注释： 使用**三个单引号** 或 **三个双引号**的方式编写
    3. 文件编码注释

## 06-数据类型

>Python中：变量无类型，数据有类型

| 类型名称 | 英文名 | 举例 | 备注 |
| --------| ------|------|------|
| 字符串 | string | 'Nancy' | 用引号包起来 |
| 整型 | int | 21 | 不带小数点的数 |
| 浮点型 | float | 47.8 | 带小数点的数 |
| …… | …… | …… | …… |

```python
print(type('Nancy'))
print(type(21))
print(type(47.8))
```

## 07-整型

- 整型：没有小数点的数字

- 当数很大时，可以使用下划线讲数字进行分组，来让数字变得更易读 `salary = 300_000`

- Python中整数的上限值，取决于执行代码的计算机的内存和处理能力

- 解除print位数限制 `sys.set_int_max_str_digits(0)`

## 08-浮点型

- 浮点数：带有小数点的数字

- 浮点数的科学计数法表示：
    - `speed_of_sound = 3.4e+2` (3.4乘以10的2次方)
    - `1e-3` (1乘以10的-3次方)
